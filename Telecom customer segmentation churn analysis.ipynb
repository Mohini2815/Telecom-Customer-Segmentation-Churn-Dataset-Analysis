**RandomUnderSampling**

We handled class imbalance of Churn category by applying RandomUnderSampler.

```python
from imblearn.under_sampling import RandomUnderSampler

# Separate features (X) and target (y)
X = df.drop('Churn', axis=1)
y = df['Churn']

# Initialize RandomUnderSampler
rus = RandomUnderSampler(random_state=42)

# Fit and apply the resampler to the data
X_resampled, y_resampled = rus.fit_resample(X, y)

# Convert the resampled data back to a DataFrame
downsample_df = pd.concat([pd.DataFrame(X_resampled, columns=X.columns), pd.DataFrame(y_resampled, columns=['Churn'])], axis=1)

downsample_df.head()
```

```python
# finding Inter quartile range for handling outliers 
percentile25 = df['Total Charges'].quantile(0.25)
percentile75 = df['Total Charges'].quantile(0.75)
```

```python
percentile25
```

```python
percentile75
```

```python
# q3 = percentile75 and q1 = percentile25
iqr = percentile75 - percentile25
```

```python
iqr
```

```python
upperlimit = percentile75 + 1.5 *iqr
lowerlimit = percentile25 - 1.5 *iqr
```

```python
upperlimit
```

```python
lowerlimit
```

```python
df[df['Total Charges']>upperlimit]
```

```python
df[df['Total Charges']<lowerlimit]
```

```python
# trimming - delete the outlier data
new_telecom_df = df.loc[(df['Total Charges']<upperlimit) & (df['Total Charges']>lowerlimit)]
print('before removing outliers',len(df))
print('after removing outliers',len(new_telecom_df))
print('outliers are',len(df)-len(new_telecom_df))
```

```python
from sklearn.preprocessing import LabelEncoder
categorical_columns =["Gender", "Contract Type", "Has Internet Service", "Has Online Security Service", "Has Device Protection Plan","Region"]

le= LabelEncoder()


for col in categorical_columns:
    new_telecom_df[col] = le.fit_transform(new_telecom_df[col])
```

```python
corr_matrix = new_telecom_df.corr()

plt.figure(figsize=(12,10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap") 
plt.show()  
```

```python
from scipy.stats import skew, kurtosis
# apply kurtosis on numerical columns
num_col = new_telecom_df.select_dtypes(include=['int64','float64']).columns
skewness = new_telecom_df[num_col].apply(skew).rename("Skewness")
value = new_telecom_df[num_col].apply(kurtosis).rename("Kurtosis")
metrics = pd.concat([skewness,value],axis=1)
```

```python
metrics
```

```python
from scipy.stats import f_oneway 
categorical_features = ["Gender", "Contract Type", "Has Internet Service", "Has Online Security Service", "Has Device Protection Plan","Region"]

for feature in categorical_features:
    groups = [new_telecom_df[new_telecom_df[feature] == category]['Churn'] for category in new_telecom_df[feature].unique()]
    f_stat, p_value = f_oneway(*groups)
    
    print(f"ANOVA results for {feature}:")
    print(f" F-statistic: {f_stat:.2f}")
    print(f" P-value: {p_value:.4f}\n")
```

```python
x = new_telecom_df.drop(columns=['Churn'])
y = new_telecom_df['Churn']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,accuracy_score


rf=RandomForestClassifier(random_state=42)
rf.fit(x_train,y_train)

y_pred_rf=rf.predict(x_test)

accuracy = accuracy_score(y_test,y_pred_rf)
print("Accuracy:",accuracy)

class_report = classification_report(y_test,y_pred_rf)
print("Classification Report:",class_report)
```

```python
from sklearn import metrics

confusion_matrix=metrics.confusion_matrix(y_test,y_pred_rf)

cm_display=metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_matrix,display_labels=[0,1])
cm_display.plot()
plt.show()
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score,classification_report

gbc_classifier=GradientBoostingClassifier(n_estimators=100,random_state=42)

gbc_classifier.fit(x_train,y_train)

y_pred_gbc=gbc_classifier.predict(x_test)

accuracy=accuracy_score(y_test,y_pred_gbc)
print("Accuracy",accuracy)

class_report=classification_report(y_test,y_pred_gbc)
print("Gradient boosting classifier's classification report\n:",class_report)
```

```python
from sklearn import metrics

confusion_matrix=metrics.confusion_matrix(y_test,y_pred_gbc)

cm_display=metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_matrix,display_labels=[0,1])
cm_display.plot()
plt.show()
```

```python
from  sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,classification_report
#initialize 
knn_classifier=KNeighborsClassifier()

knn_classifier.fit(x_train,y_train)

y_pred_knn=knn_classifier.predict(x_test)

accuracy=accuracy_score(y_test,y_pred_knn)
print("Accuracy",accuracy)

classification_rep=classification_report(y_test,y_pred_knn)
print("knn classifier's classification report\n:",classification_rep)
```

```python
from sklearn import metrics

confusion_matrix=metrics.confusion_matrix(y_test,y_pred_knn)

cm_display=metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_matrix,display_labels=[0,1])
cm_display.plot()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score


model=LinearRegression()
model.fit(x_train, y_train)
y_pred_test=model.predict(x_test)#for evaluating test set


#Evaluation Matrix
#now we are saving predictions
lr_metrics={'Mean Squared Error ':mean_squared_error(y_test,y_pred_test),
           'Root Mean Squared Error':np.sqrt(mean_squared_error(y_test,y_pred_test)),
           'Mean Absolute Error':mean_absolute_error(y_test,y_pred_test),
           'R Squared':r2_score(y_test,y_pred_test)}
print(lr_metrics)
```
